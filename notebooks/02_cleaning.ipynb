```python
import pandas as pd
import sqlite3

from jobanalysis.paths import RAW_DATA
```

```python
conn = sqlite3.connect(RAW_DATA)

df = pd.read_sql_query("SELECT * FROM vacancies", conn)
conn.close()

df.head()
```

```python
df["salary"] = df["salary"].astype(int)
df["salary"] = df["salary"].replace(0, float("nan"))
```

```python
df["created_at"] = pd.to_datetime(df["created_at"])
df["deadline_at"] = pd.to_datetime(df["deadline_at"])
df = df.dropna(subset="deadline_at")
```

```python
df["company"] = df["company"].replace("Company", None)
```

```python
from jobanalysis.clean import parse_view_count

df["view_count"] = df["view_count"].apply(parse_view_count)
```

```python
df = df.drop(columns=["id", "company_id", "url", "fetched_at", "direct_apply"])
```

```python
from jobanalysis.paths import PROCESSED_DATA

conn = sqlite3.connect(PROCESSED_DATA)
df.to_sql("vacancies", conn, if_exists="replace", index=False)
conn.close()
```
